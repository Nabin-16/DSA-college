```python
G = {
    's':{'t':10, 'y':5},
    't':{'x':1, 'y':2},
    'x':{'z':4},
    'z':{'s':7, 'x':6},
    'y':{'x':9, 't':3}
}
```

```python
import math
def INITIALIZE_SINGLE_SOURCE(G, s):
  cost = dict()
  prev = dict()
  for vertex in G.keys():
   cost[vertex]= math.inf
   prev[vertex]=""
  cost[s]= 0
  return cost,prev
```

```python
def RELAX(G, u , v, cost, prev):
    if cost[v]>cost[u] + G[u][v]:
        cost[v]= cost[u] +G[u][v]
        prev[v]= u
    return cost, prev
```

```python
from queue import PriorityQueue
def DJ(G, s):
   cost, prev = INITIALIZE_SINGLE_SOURCE(G, s)
   PQ= PriorityQueue()
   for vertex in G.keys():
    PQ.put((cost[vertex], vertex))
   visited = set()
   while not PQ.empty():
        current_cost, currentVertex = PQ.get()
        if currentVertex in visited:
            continue
        visited.add(currentVertex)
        for neighbour in G[currentVertex]:
            if neighbour not in visited:
                old_cost = cost[neighbour]
                cost, prev = RELAX(G, currentVertex, neighbour, cost, prev)
                if cost[neighbour] < old_cost:
                    PQ.put((cost[neighbour], neighbour))
   return cost, prev
```

```python
s ='s'
cost,prev =DJ(G,s)
print(cost)
print(cost)
```

```python
def RECONSTRUCT_PATH(G, vertex, prev):
    path= vertex
    while prev[vertex] !="":
        path = prev[vertex]+'->'+path
        vertex = prev[vertex]
    return path
s='s'
cost, prev= DJ(G,s)
for vertex in G.keys():
    print(f'shortest path from {s} to {vertex} is RECONSTRUCT_PATH {vertex, prev}')
    print(f'Cost is {cost[vertex]}')
```
